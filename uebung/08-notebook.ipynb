### *Dependenzanalyse:*

```python
import spacy
analyze = spacy.load('en_core_web_sm')

from spacy import displacy
options = {"compact": True, "collapse_phrases": True, "distance" : 100, "arrow_stroke":1}

def printConll(doc):
    print("CONLL:")
    for i, word in enumerate(doc):
        if word.head == word:
            head_idx = 0
        else:
            head_idx = word.head.i + 1
        print(
            "%d\t%s\t%s\t%s\t%s\t%s\t%s" % (
                i+1,
                word,
                word.lemma_,
                word.tag_,
                word.ent_type_,
                str(head_idx),
                word.dep_
            )
        )
```

```python
analyzed = analyze(sentence)
displacy.render(analyzed)
printConll(analyzed)
```
